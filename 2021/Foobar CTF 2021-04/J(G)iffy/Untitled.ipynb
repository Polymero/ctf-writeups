```python
from Crypto.Util.number import long_to_bytes,bytes_to_long
```

```python
with open("chall",'rb') as f:
    
    hx = f.read()
    
    f.close()
    
data = hx[:6]
sign = b''
for i in [0x47, 0x49, 0x46, 0x38, 0x39, 0x61]:
    sign += long_to_bytes(i)
```

```python
xorkey = long_to_bytes(bytes_to_long(data) ^ bytes_to_long(sign))
orkey = long_to_bytes(bytes_to_long(data) | bytes_to_long(sign))
print(xorkey); print(orkey)
```

```python
with open("chall_xor",'wb') as f:
    
    new_file = b''
    for i,b in enumerate(hx):
        new_file += long_to_bytes(b ^ xorkey[i % len(xorkey)])
        
    f.write(new_file)
    
    f.close()
```

```python
with open("chall_or",'wb') as f:
    
    new_file = b''
    for i,b in enumerate(hx):
        new_file += long_to_bytes(b | orkey[i % len(xorkey)])
        
    f.write(new_file)
    
    f.close()
```
