```python
from Crypto.Util.number import long_to_bytes
```

```python
with open('baby_double_xor.enc','rb') as f:
    
    enc = f.read()#[:32]
    print(enc)
    
    f.close()

keylen = 0

for i in range(256**keylen):
    
    key = long_to_bytes(i)
    while len(key) < keylen:
        key = b'\x00' + key
        
    assert len(key) == keylen
        
    dec = b''
    
    for j,c in enumerate(enc):
        
        dec += long_to_bytes( c ^ key[j % keylen] )
    

    print(dec)
```

```python
encHS = b'\n\x8b\xde`'

pngHS = [0x89, 0x50, 0x4E, 0x47]
pngxor = bytes([encHS[i] ^ pngHS[i] for i in range(0,4)])

pngdec = b''
for i,c in enumerate(enc):
    pngdec += long_to_bytes( c ^ pngxor[i % len(pngxor)] )
```
