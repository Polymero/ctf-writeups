```python
# Primes
p, q = chall.genPrimes(1024)
n = p*q
phi = (p-1)*(q-1)

# MSB of n
nmsb = int(bin(n)[2:2+256*2],2) << 6*256; print(nmsb)

# Square root the MSB of n
offset=0
nrootmsb = bin(gmpy2.iroot(nmsb,2)[0])[2:2+256*2-offset]
pmsb = bin(p)[2:2+256*2-offset]

# Check similarity
while nrootmsb != pmsb:
    offset += 1
    nrootmsb = bin(gmpy2.iroot(nmsb,2)[0])[2:2+256*2-offset]
    pmsb = bin(p)[2:2+256*2-offset]
print('Offset:', offset)
print()

# Create leaked primes
pleak = pmsb + offset*'?' + 256*'0' + 256*'?'
qleak = pleak

# Visual checks
assert nrootmsb == pmsb
assert len(pleak) == 4*256
print('P_leak:')
print(pleak[:256])
print(pleak[256:2*256])
print(pleak[2*256:3*256])
print(pleak[-256:])
print()
print('Leak:', len(pleak.replace('?',''))/len(pleak)*100, '%')
```

```python
nmsb = bin((int(nrootmsb,2) << 512+offset)**2)[2:]
print(len(nmsb))
nmrlsb = bin(n - int(nmsb,2))[-512:]; print(len(nmrlsb), nmrlsb)
print(int(nmrlsb,2))
```

```python
plsb = int(bin(p)[2+512+256:],2); print(plsb)
qlsb = int(bin(q)[2+512+256:],2); print(qlsb)
assert plsb*qlsb == int(nmrlsb,2)
```
