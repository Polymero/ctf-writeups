```python
# Imports
from math import factorial as fac
import chall
import gmpy2
```

```python
def binom(n, r):
    return fac(n) // (fac(r) * fac(n-r))

def v(m, e):
    if e == 0:
        return 2
    if e == 1:
        return m
    return(m*v(m, e-1) - v(m, e-2)) 

def myv(m, e):
    k = e//2 - 1
    ans = m**e
    for i in range(k+1):
        ans += e * binom(k, i) * (-1)**(i+1) * m**(e-2*(i+1))
    return ans
```

```python
# Primes
p, q = chall.genPrimes(1024)
n = p*q
phi = (p-1)*(q-1)

# MSB of n
nmsb = int(bin(n)[2:2+256*2],2) << 6*256; print(nmsb)

# Square root the MSB of n
offset=0
nrootmsb = bin(gmpy2.iroot(nmsb,2)[0])[2:2+256*2-offset]
pmsb = bin(p)[2:2+256*2-offset]

# Check similarity
while nrootmsb != pmsb:
    offset += 1
    nrootmsb = bin(gmpy2.iroot(nmsb,2)[0])[2:2+256*2-offset]
    pmsb = bin(p)[2:2+256*2-offset]
print('Offset:', offset)
print()

# Create leaked primes
pleak = pmsb + offset*'?' + 256*'0' + 256*'?'
qleak = pleak

# Visual checks
assert len(pleak) == 4*256
print('P_leak:')
print(pleak[:256])
print(pleak[256:2*256])
print(pleak[2*256:3*256])
print(pleak[-256:])
print()
print('Leak:', len(pleak.replace('?',''))/len(pleak)*100, '%')
```

```python
ofs = 0

while bin(n - (int(nrootmsb,2) << (2*256+offset))**2)[-1024-ofs:] == bin(p*q)[2+1024-ofs:]:
    ofs += 1
ofs -= 1
print(ofs)

nmr_msb = bin(n - (int(nrootmsb,2) << (2*256+offset))**2)[-1024-ofs:]

print(bin(n - (int(nrootmsb,2) << (2*256+offset))**2)[-1024-ofs:])
print(bin(p*q)[2+1024-ofs:])
print(len(bin(p*q)[2+1024-ofs:]))
```
